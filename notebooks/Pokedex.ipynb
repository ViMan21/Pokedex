```python
from fastai.vision import *
from fastai.widgets import *
```

```python
path = Path("storage/")
fileName = 'labels.txt'
pokemonList = [line.rstrip('\n') for line in open(path/fileName)]
    
pokemonList
```

```python
path = Path('storage/data/poke')
source = Path('storage/urls')
```

```python
for pokemon in pokemonList:
    folder = pokemon
    file = pokemon + '.txt'
    dest = path/folder
    dest.mkdir(parents=True, exist_ok=True)
    print("Downloading {} images...".format(pokemon))
    download_images(source/file, dest, max_pics=300)
```

```python
path.ls()
```

```python
classes = pokemonList
```

```python
for c in classes:
    print(c)
    verify_images(path/c, max_size=500)
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=5, figsize=(7,8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```python
learn.fit_one_cycle(8)
```

```python
learn.save('stage-1')
```

```python
learn.export()
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-5,1e-4))
```

```python
learn.save('stage-2')
```

```python
learn.export()
```

```python
doc(interp.plot_top_losses)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix(figsize=(30,30), dpi=60)
```

```python
interp.most_confused()
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```
